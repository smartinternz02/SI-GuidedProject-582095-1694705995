```python
#sai yaswanth reddy suram
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the 'car_crashes' dataset from seaborn
df = sns.load_dataset('car_crashes')
```

```python
# Display basic information about the dataset
df.info()
```

```python
# Display the first 5 rows of the dataset
df.head()
```

```python
# Display the last 5 rows of the dataset
df.tail()
```

```python
# Display summary statistics for numerical columns in the dataset
df.describe()
```

```python
# Count the occurrences of each value in the 'speeding' column
# Inference: This counts the occurrences of each value in the speeding-related accidents column.
df["speeding"].value_counts()
```

```python
# Count the occurrences of each value in the 'total' column
# Inference: This counts the occurrences of each value in the total number of accidents column.
df["total"].value_counts()
```

```python
# Count the occurrences of each value in the 'ins_losses' column
# Inference: This counts the occurrences of each value in the insurance losses column.
df["ins_losses"].value_counts()
```

```python
# Create a scatterplot of 'ins_premium' vs. 'alcohol'
# Inference: This scatterplot can help explore the relationship between insurance premiums and alcohol-related accidents.
sns.scatterplot(x="ins_premium", y="alcohol", data=df)
```

```python
# Create a scatterplot of 'total' vs. 'speeding'
# Inference: This scatterplot can help examine the correlation between the total number of accidents and accidents related to speeding.
sns.scatterplot(x="total", y="speeding", data=df)
```

```python
# Create a scatterplot of 'speeding' vs. 'ins_losses'
# Inference: This scatterplot can help explore the relationship between speeding-related accidents and insurance losses.
sns.scatterplot(x="speeding", y="ins_losses", data=df,color="pink")
```

```python
# Create a scatterplot of 'ins_losses' vs. 'ins_premium'
# Inference: This scatterplot explores the relationship between insurance losses and insurance premiums.
sns.scatterplot(x="ins_losses", y="ins_premium", data=df,color="purple")
plt.xlabel("Insurance Losses")
plt.ylabel("Insurance Premium")
plt.title("Scatterplot of Insurance Losses vs. Insurance Premium")
plt.show()
```

```python
# Create a scatterplot matrix of selected columns
selected_columns = df[['total', 'ins_losses', 'alcohol', 'speeding']]
pd.plotting.scatter_matrix(selected_columns, figsize=(8, 8),color="green")
plt.suptitle("Scatterplot Matrix of Selected Columns")
plt.show()
```

```python
# Create a lineplot of 'total' vs. 'speeding' without confidence intervals
# Inference: This lineplot shows the trend in speeding-related accidents as the total number of accidents increases.
sns.lineplot(x="total", y="speeding", data=df, ci=None)
```

```python
#Inference: INVERSLY PROPORTIONAL
#As the percentage of accidents where drivers were not distracted (x-axis) increases, the total number of accidents (y-axis)
#tends to decrease. This implies that a lower rate of distraction among drivers is associated with fewer accidents.
sns.lineplot(x="not_distracted",y="total",data=df,color='green')
```

```python
#INFERENCE: CORRELATION EXSISTS
#As alcohol consumption increases, speeding tends to increase as well, indicating a positive or direct correlation
#between the two variables.
sns.lineplot(x="alcohol",y="speeding",data=df,color='blue')
```

```python
# Create a distribution plot for the 'total' column
# Inference: This plot illustrates the distribution of the total number of accidents in the dataset.
sns.distplot(df["total"],color="red")
```

```python
#INFERENCE:
#This plot illustrates the distribution of speeding-related accidents in the dataset.
#The distribution of "speeding" incidents is right-skewed, with a majority of regions having lower speeding rates.
sns.distplot(df["speeding"])
plt.show()
```

```python
# Create a distribution plot for the 'ins_losses' column
# Inference: This plot illustrates the distribution of insurance losses in the dataset.
sns.distplot(df["ins_losses"],color="orange")
```

```python
#Inference: As the speeding increases the rate of total crashes also increases, but there is a drasting increase in the creashes if there is a high alcohol rate along with speeding.
#multi  variate
sns.relplot(x="speeding",y="total",data=df,hue="alcohol")
```

```python
# Create a relational plot (scatterplot) of 'total' vs. 'ins_losses' with hue 'no_previous'
# Inference: This plot explores the relationship between total accidents, insurance losses, and whether there were previous accidents.
sns.relplot(x="total", y="ins_losses", data=df, hue="no_previous")
```

```python
#INFERENCE:The relational plot illustrates a potential association between higher alcohol involvement and increased speeding incidents. Variations in insurance premium costs suggest that regions with elevated alcohol involvement may face diverse insurance pricing.
sns.relplot(x="alcohol", y="speeding", hue="ins_premium", data=df)
plt.title("Relationship between Alcohol and Speeding with Insurance Premium")
plt.xlabel("Alcohol Involvement (%)")
plt.ylabel("Speeding Incidents")
plt.show()
```

```python
# Create a bar plot of 'abbrev' vs. 'speeding'
# Inference: This bar plot displays the average speeding incidents by abbreviation (likely representing states or regions).
sns.barplot(data=df, x="abbrev", y="speeding")
```

```python
#Inference: INVERSLY PROPORTIONAL
#As the total number of accidents (x-axis) increases, the percentage of accidents involving drivers with no previous accidents (y-axis) tends to decrease. This implies that a higher accident rate is associated with a lower percentage of inexperienced drivers.
#BAR PLOT
plt.subplots(figsize=(20,8))
sns.barplot(data=df,x="total",y="no_previous",ci=None)
plt.show()
```

```python
# Create a histogram of the 'speeding' column
# Inference: This histogram visualizes the distribution of speeding-related accidents.
plt.hist(x="speeding")
plt.show()
```

```python
# Create a histogram of the 'speeding' column
# Inference: This histogram visualizes the distribution of speeding-related accidents.
plt.hist(x="speeding",color="red")
plt.show()
```

```python
# Create a joint plot of 'total' vs. 'speeding'
# Inference: This joint plot combines a scatterplot of 'total' vs. 'speeding' with marginal histograms.
sns.jointplot(x="total", y="speeding", data=df,color="yellow")
```

```python
# Create a joint plot of 'alcohol' vs. 'ins_losses'
# Inference: This joint plot combines a scatterplot of 'alcohol' vs. 'ins_losses' with marginal histograms.
sns.jointplot(x="alcohol", y="ins_losses", data=df,color="black")
```

```python
# Create a box plot of 'total' vs. 'alcohol'
# Inference: This box plot shows the distribution of alcohol-related accidents across different total accident values.
sns.boxplot(x="total", y="alcohol", data=df)
```

```python
# Create a box plot of 'ins_premium' for different regions (abbreviations)
# Inference: This box plot shows the distribution of insurance premiums for different regions.
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="abbrev", y="ins_premium")
plt.xlabel("Region (Abbreviation)")
plt.ylabel("Insurance Premium")
plt.title("Box Plot of Insurance Premium by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a box plot of 'total' accidents grouped by distracted driving status
plt.figure(figsize=(8, 6))
sns.boxplot(x="not_distracted", y="total", data=df)
plt.xlabel("Distracted Driving")
plt.ylabel("Total Accidents")
plt.title("Box Plot of Total Accidents by Distracted Driving Status")
plt.show()
```

```python
# Calculate the correlation matrix for the dataset
corr = df.corr()
# Display the correlation coefficients between variables in the dataset
print(corr)
```

```python
# Create a heatmap of the correlation matrix
# Inference: This heatmap visually represents the correlations between different numerical columns in the dataset.
sns.heatmap(corr, annot=True)
```

```python
# Create a regplot of 'total' vs. 'ins_losses' with a regression line
# Inference: This regplot shows the linear relationship between total accidents and insurance losses.
sns.regplot(x="total", y="ins_losses", data=df,color="red")
```

```python
# Create a regression plot of 'ins_premium' vs. 'ins_losses' with a regression line
# Inference: This plot shows the linear relationship between insurance premiums and insurance losses.
sns.regplot(x="ins_premium", y="ins_losses", data=df,color="green")
plt.xlabel("Insurance Premium")
plt.ylabel("Insurance Losses")
plt.title("Regression Plot of Insurance Premium vs. Insurance Losses")
plt.show()
```

```python
# Create a histogram of 'total' with custom bins
# Inference: This histogram provides a different view of the distribution of total accidents with custom bin sizes.
plt.hist(df["total"], bins=[0, 5, 10, 15, 20, 25, 30, 35, 40],color="brown")
plt.xlabel("Total Accidents")
plt.ylabel("Frequency")
plt.title("Histogram of Total Accidents with Custom Bins")
plt.show()
```
